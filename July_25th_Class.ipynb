```
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
%matplotlib inline
import seaborn as sns
sns.set()
```

```
df = pd.read_csv('https://raw.githubusercontent.com/nytimes/covid-19-data/master/us-states.csv')
```

```
df['date']=pd.to_datetime(df['date'])
```

```
df.tail(10)
```

```
df.pivot(index='date',columns='state', values='deaths').tail(7)
```

```
df.groupby('state').mean()
```

```
df = df.reindex(sorted(df.columns), axis=1)
df.tail(10)
```

```
df.tail(30)
```

```
Texas_deaths = df[df['state']=='Texas']
Texas_deaths.head()
```

```
Texas_deaths.shape
```

```
newdeaths=[]
previous_deaths = 0
for index, row in Texas_deaths.iterrows():
  current_deaths = row ['deaths']
  newdeaths_count = current_deaths - previous_deaths
  previous_deaths = current_deaths  
  newdeaths.append(newdeaths_count)
```

```
Texas_deaths['new_deaths'] = newdeaths
```

```
Texas_deaths.tail()
```

```
Texas_deaths['date']=pd.to_datetime(Texas_deaths['date'])
```

```
sns.distplot(Texas_deaths['new_deaths'])
```

```
sns.jointplot(x='date', y='new_deaths', data=Texas_deaths)
```

```
texas_cases = Texas_deaths['cases'][-11:-1]
texas_deaths = Texas_deaths['deaths'][-11:-1]
texas_dates = Texas_deaths['date'][-11:-1]
texas_new_deaths = Texas_deaths['new_deaths'][-11:-1]
```

```
Texas_deaths.plot(kind='bar')
```

```
plt.figure(figsize=(15,7))
plt.bar(x=texas_dates, height=texas_new_deaths, color='purple')
plt.xticks(rotation=90)
plt.show()
```

```
states = ['California','Florida', 'Texas','Arizona','Georgia']
states
```

```
plt.figure(figsize=(15,7))
for state in states:
  df_st = df[(df['state']==state) & (df['date']>='2020-07-16')]
  plt.plot(df_st['date'],
           df_st['cases'],
           linewidth=2
           )
  plt.xticks(rotation=90)
plt.legend(states)
plt.show()
```
